Данная статья посвящена сравнению двух методов проверки статистических гипотез, основанных на частотном и байесовском подходах. Данное изложение представляет собой неполный конспект этой статьи, дополненный моими собственными рассуждениями.

### Формула Байеса

Начнем с формулы Байеса:

$P(AB) = P(A|B)P(B) = P(B|A)P(A)$

Собственно формула Байеса - это второе равенство. $В$ такой записи формула выглядит очевидной: вероятность наступления сразу обоих событий ($A$ и $B$) равна вероятности события $B$, помноженному на вероятность наступления $A$ при $B$, и она же равна вероятности события $A$, помноженному на вероятность наступления $B$ при $A$. Поэтому эти две величины равны между собой. Если выразить отсюда $P(A|B)$, то получим классическую форму записи формулы Байеса:

$P(A|B) = \cfrac{P(B|A)P(A)}{P(B)}$

В частотном подходе сама формула Байеса не оспаривается. Вопрос лишь в том, до какой степени ее допустимо применять, а точнее - *можно ли подставлять вместо A или B не только события, но и гипотезы*? Частотный подход говорит, что нельзя. Байесовский подход говорит, что можно. Почему же они не могут договориться между собой? Потому что в них по-разному определяется понятие *вероятности*.

### Частотный и байесовский подход

В **частотном подходе** (фреквентизм, или фишеровская статистика) вероятность - это частота возникновения некоего события при бесконечном повторении эксперимента. Согласно этому определению, некорректно говорить о *вероятности гипотезы*. Гипотеза либо верна, либо нет. Нам может быть неизвестно верна ли она, поэтому мы можем говорить о степени уверенности в верности гипотезы. Но применять к гипотезе понятие вероятности нельзя, а значит *нельзя подставлять гипотезы в формулу Байеса*.

В **байесовском подходе** вероятность - это степень нашей субъективной уверенности в истинности какого-либо суждения. Такая более широкая трактовка дает нам право говорить о вероятности гипотезы, а значит *можно подставлять гипотезы в формулу Байеса*.

### Байесовский вывод

В байесовском статистическом выводе формула Байеса применяется следующим образом:

$P(\mathit{Hypotesis}|\mathit{Data}) = \cfrac{P(\mathit{Data}|\mathit{Hypotesis})P(\mathit{Hypotesis})}{P(\mathit{Data})}$

Данная формула отражает то, как меняется наша уверенность в некой гипотезе при получении новой порции экспериментальных данных. Используется следующая терминология:

$P(\mathit{Hypotesis})$ - **априорная вероятность гипотезы (англ. *prior*)**. Степень нашей уверенности в истинности гипотезы до эксперимента. Происходит от латинского *"a priori"* - "от предшествующего". 

$P(\mathit{Hypotesis}|\mathit{Data})$ - **апостериорная вероятность гипотезы (англ. *posterior*)**. Степень нашей уверенности в истинности гипотезы после эксперимента. Происходит от латинского *"a posteriori"* - "из последующего".

$P(\mathit{Data}|\mathit{Hypotesis})$ - **правдоподобие (англ. *likelihood*)** данных при условии истинности гипотезы, то есть вероятность получить такие экспериментальные данные, если гипотеза верна.

$P(\mathit{Data})$ - **частная вероятность $\mathit{Data}$, англ. *marginal probability, или evidence***. Вероятность получить такие экспериментальные данные. Эту величину можно вычислить по формуле, перебрав все возможные гипотезы, их вероятности и условные вероятности получить при них такие данные:

$P(\mathit{Data}) = \sum_{H_i} P(\mathit{Data}|H_i)P(H_i)$

Что значит "все возможные гипотезы"? Это обычно означает все возможные значения параметра, который мы хотим измерить. Рассмотрим на примере.

### Пример байесовского вывода

Пусть мы работаем с выборкой данных, и наша задача - оценить матожидание ($\mu$) распределения, из которого эти данные получены. Пусть при этом известно, что распределение нормально со стандартным отклонением 30. Автор статьи - астрофизик, и его пример описывает звезду, яркость которой мы измеряем. Выборка - это набор измеренных значений яркости, все они разные из-за погрешности измерений.

Каждое конкретное значение матожидания будет гипотезой, а все гипотезы - все возможные значения матожидания. Количество таких гипотез равно количеству действительных чисел (континуум). Формулу Байеса мы применяем ко всем возможным гипотезам (значениям неизвестного параметра), это выглядит таким образом:

$for X \in (-\infty, +\infty): P(\mathit{\mu = X}|\mathit{Data}) = \cfrac{P(\mathit{Data}|\mathit{\mu = X})P(\mathit{\mu = X})}{P(\mathit{Data})}$

Конечно вероятность того, что $\mu$ в точности равна какому-то числу, равна нулю, поэтому вместо вероятности используется плотность вероятности. Если мы внимательно посмотрим на записанное выше выражение, то увидим, что в нем одна функция умножается на другую, делится на некое число, и в результате получается функцию.

Поэтому мы вернемся к исходной записи, но теперь под априорной и апостериорной вероятностью понимаем плотность вероятности:

$P(\mathit{H}|\mathit{Data}) = \cfrac{P(\mathit{Data}|\mathit{H})P(\mathit{H})}{P(\mathit{Data})}$

$P(\mathit{H})$ и $P(\mathit{H}|\mathit{Data})$ - априорная и апостериорная плотности вероятности интересующего нас параметра, то есть наши знания до и после эксперимента.

Заметим, что $P(\mathit{Data})$ не зависит от значения параметра $\mu$, то есть его можно считать константой. Апостериорная плотность вероятности - это распределение, то есть его интеграл обязан быть равен единице. Поэтому $P(\mathit{Data})$ играет роль коэффициента нормализации - он ровно такой, что интеграл всего выражения равен единице. Поэтому формулу можно переписать таким образом:

$P(\mathit{H}|\mathit{Data}) = \mathit{Normalize} \big[ P(\mathit{Data}|\mathit{H})P(\mathit{H}) \big]$

Здесь операция $\mathit{Normalize}$ означает L1-нормализацию по всем возможным $H$.

В итоге мы максимально упростили формулу: мы просто перемножаем две функции и получаем третью.

### Что брать в качестве prior?

В примере выше мы ничего заранее не знаем о матожидании. Непонятно, что в этом случае считать априорной вероятностью. На практике обычно берут так называемый flat prior - равномерное распределение на некоем очень большом отрезке. Непонятно какой отрезок брать, поэтому можно рассмотреть предельный случай, когда границы отрезка стремятся к бесконечности, а плотность вероятности, соответственно, стремится к нулю. Но заметим, что результат мы все равно нормализуем, поэтому умножение $P(H)$ на некую константу не повлияет на результат. В итоге за $P(H)$ берут функцию, тождественно равную единице. Этот прием называется "improper prior", потому что при этом интеграл $P(H)$ бесконечен, а значит $P(H)$ не является распределением.

### Считаем апостериорную вероятность

Нам осталось посчитать $P(\mathit{Data}|\mathit{H})$ и нормализовать результат. По смыслу это выражение означает вероятность (точнее плотность вероятности) получить текущие экспериментальные данные, при всех возможных значениях параметра. Поскольку нам известно, что распределение нормально со стандартным отклонением 30, то из значения $\mu$ можно непосредственно вывести правдоподобие:

$\mathcal{L}(Data|mu = X) = \prod_{i=1}^N \mathcal{N}_{\mu, 30}(Data_i)$

Здесь $Data_i$ означает $i$-е наблюдение в выборке, а $\mathcal{N}_{\mu, 30}(Data_i)$ означает функцию плотности вероятности нормального распределения с параметрами $\mu, 30$, взятую в точке $Data_i$.

Чтобы получить апостериорную плотность вероятности для значений параметра $\mu$, описанное выше произведение мы считаем для всех значений $\mu$, а затем нормалируем полученный график.

### Максимизация правдоподобия

Часто задачей является не построить указанный график целиком, а лишь найти его максимум, то есть найти такое значение параметра $\mu$, при котором текущие экспериментальные данные наиболее вероятны (среди всех значений $\mu$). Такая задача называется **максимизацией правдоподобия**.

В этом случае задача упрощается. В этом нам удобнее считать не само правдоподобие, $\mathcal{L}(Data|mu = X)$, а его логарифм, поскольку в этом случае произведение превращаемся в сумму. Затем мы ищем такое $\mu$, при котором эта сумма достигает максимума:

$\mu_\mathit{most\ probable} = argmax_\mu \sum_{i=1}^N \log \mathcal{N}_{\mu, 30}(Data_i)$

Задача максимизации правдоподобия не содержит вероятностей гипотез, а поэтому применяется также и в частотном выводе. В байесовском выводе максимизация правдоподобия представляет собой частный случай оценки апостериорного максимума, когда априорное распределение равномерно. Вспомним, что в байесовском выводе мы умножали две функции, одна из которых была константой, а вторая как раз и была функцией правдоподобия. В итоге мы получали апостериорную вероятность, и могли найти ее максимум, то есть наиболее вероятную гипотезу.

### Оценка погрешности

Углубившись в байесовский вывод, мы до сих пор ничего не говорили о частотном (фреквентистском) статистическом выводе. Этот вывод часто сводится к максимизации функции правдоподобия. Однако часто также требуется оценить допустимую погрешность. И здесь выясняется, что частотный и байесовский вывод делают это существенно по-разному.

- При оценке некоего параметра в байесовском подходе ищется некий регион (credible region), в простом случае - интервал, в котором параметр лежит с некой вероятностью. Например, $\mu$ лежит в интервале от 1 до 2 с верояностью 95%.
- При оценке некоего параметра в частотном подходе мы не можем говорить о вероятности нахождения параметра в заданном регионе, так как это гипотеза, а вероятность неприменима к гипотезам. Поэтому ответ ищется в виде следующей формулировки: если эксперимент будет повторяться много раз, то в 95% случаев найденный интервал (confidence interval) будет содержать истинное $\mu$. При этом интервал в каждой итерации эксперимента будет разным, но будет искаться по одним и тем же правилам.

Формулировка, применяемая в частотном подходе, говорит о том, что в каких-то случаях мы можем получить в принципе невозможный интервал, но вероятность этого невелика. В научной литературе часто встречается формулировка, что с 95% вероятностью неизвестный параметр лежит в указанном диапазоне. Автор статьи утверждает, что это является неверным применением байесовской интерпретации к интервалу, полученному фреквентистским подходом. Автор статьи также приводит пример задачи, в которой фреквентистский подход дает такой доверительный интервал, в котором искомый параметр никак не может лежать (см. раздел "Truncated Exponential: A Frequentist Approach").

### Заключение: сравнение подходов

Байесовский подход позволяет применять понятие вероятности к гипотезам, однако при этом требует введения априорных гипотез (priors). Это означает, что в байесовском подходе мы должны "отткалкиваться" от каких-то предположений и не можем строить доказательное знание на пустом месте. Даже неинформативный "flat prior" может нести в себе информацию (согласно автору, со ссылкой на статью "Inverse problems as statistics").

> A frequentist will point out that the prior is problematic
when no true prior information is available. Though it might
seem straightforward to use an **uninformative prior** like
the flat prior mentioned above, there are some surprising
subtleties involved.<br/>
The flat prior in this case can be motivated by maximum entropy. Still, the use of uninformative priors like this often
raises eyebrows among frequentists: there are good arguments that even
“uninformative” priors can add information.<br/>
It turns out that in many situations, a
truly uninformative prior cannot exist! Frequentists point out
that the subjective choice of a prior which necessarily biases
the result should have no place in scientific data analysis.<br/>
A Bayesian would counter that frequentism doesn’t solve
this problem, but simply skirts the question. **Frequentism can
often be viewed as simply a special case of the Bayesian
approach for some (implicit) choice of the prior**: a Bayesian
would say that it’s better to make this implicit choice explicit,
even if the choice might include some subjectivity. Furthermore, as we will see below, the question frequentism answers
is not always the question the researcher wants to ask.

И далее:

> ...we must be aware that a **flat prior is not
uninformative**. Because of the nature of the slope, a flat prior
leads to a much higher probability for steeper slopes. One might imagine addressing this by transforming variables ... .

Таким образом, допуская применение формулы Байеса к гипотезам, байесионист накладывает на себя дополнительную ответственность за выбор априорной гипотезы, могущей повлиять на весь ход статистического вывода.

Помимо "философских" различий, байесовский подход может быть вычислительно тяжелым, если требуется проверка сложных гипотез со многими параметрами.

>Though Bayesianism has some nice features in theory, in
practice it can be extremely computationally intensive: while
simple problems like those examined above lend themselves to
relatively easy analytic integration, real-life Bayesian computations often require numerical integration of high-dimensional
parameter spaces.

Для проверки сложных гипотез в байесовском подходе на практике используются такие итерационные методы, как Markov Chain Monte Carlo (MCMC). Этому методу автор отводит несколько страниц, показывая подробный пример решения задачи линейной регрессии с оценкой погрешности. При этом сравниваются оба подхода, а для реализации байесовского вывода с помощью MCMC и используется несколько python-библиотек. Здесь мы не будем разбирать этот пример (см. раздел "Application: A Simple Linear Model").

В заключение автор делает следующий вывод (перевод последнего абзаца):

>Выбор метода определяется отчасти принятой идеологией, но также и природой решаемой задачи. Частотный подход часто легче в вычислительном плане и лучше подходит для процессов, которые можно повторять бесконечно. Байесовский подход лучше подходит для задач с небольшим объемом данных и распределениями, далекими от нормальных. Частотный подход также можно работать с небольшими данными и сложными распределениями, но требует при этом высокого уровня экспертизы. Байесовский подход требует указания априорной гипотезы, которая может быть субъективной, и часто требует тяжелых вычислений с помощью метода MCMC. Концептуально байесовский подход более прямолинеен и дает более естественный ответ в форме "как текущие экспериметальные данные ограничивают неизвестные параметры исследуемой модели?". При корректном использовании оба подхода могут быть эффективно использованы для интерпретации различных научных и инженерных данных.
