В последнее время в распознавании объектов нейронными сетями достигнут существенный прогресс. Их уже сложнее сбить с толку необычными текстурами объектов, необычным фоном, и они уже не так часто принимают ледпардовый диван за леопарда. Но серьезные проблемы все еще остаются. Например, нейронная сеть OFA ([Wang et al., 2022]($Unifying Architectures, Tasks, and Modalities Through a Simple Sequence-to-Sequence Learning Framework$)) умеет выполнять много разных задач, связанных с изображениями и текстами, в том числе описывать в виде текста увиденное на изображении. Однако при этом она часто путает количество объектов или описывает несуществующие объекты.
