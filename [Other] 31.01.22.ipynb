Исходно задача ставится следующим образом ($\mathcal{L}$ - функция потерь):

$f_1, f_2, f_{12} = \text{argmin}_{f_1, f_2, f_{12}} \mathbb{E}_{x_1, x_2, y \sim D} \ \mathcal{L}(y, f_1(x_1) + f_2(x_2) + f_{12}(x_1, x_2)) \tag{1}$

То есть найти такую функцию $f_1(x_1) + f_2(x_2) + f_{12}(x_1, x_2)$, которая позволяет предсказывать $y$ наилучшим образом (с наименьшей ожидаемой функцией потерь).

Но у такой задачи много решений, так как для любых $f_1, f_2$ существует $f_{12}$ такое, что выполняется первое равенство.

Поэтому хотелось бы, чтобы $f_1, f_2$ тоже были оптимальны, то есть:

$f_1, f_2 = \text{argmin}_{f_1, f_2} \mathbb{E}_{x_1, x_2, y \sim D}\ \mathcal{L}(y, f_1(x_1) + f_2(x_2)) \tag{2}$

Попробуем суммировать функции потерь из первого и второго уравнения:

$f_1, f_2, f_{12} = \text{argmin}_{f_1, f_2, f_{12}} \mathbb{E}_{x_1, x_2, y \sim D} \Big[ \mathcal{L}(y, f_1(x_1) + f_2(x_2) + f_{12}(x_1, x_2)) + \lambda\mathcal{L}(y, f_1(x_1) + f_2(x_2)) \Big] \tag{3}$

Верно ли, что решение этой задачи всегда даст такие $f_1, f_2, f_{12}$, что уравнения $(1)$, $(2)$ тоже будут выполняться?

------

**Теорема**.

Если тройка функций $f_1, f_2, f_{12}$ удовлетворяет уравнению $(3)$, то она удовлетворяет также уравнению $(1)$.

**Доказательство**.

Пусть это не так, то есть существуют более оптимальные $f_1^\ast, f_2^\ast, f_{12}^\ast$ с точки зрения условия в уравнении $(1)$. Без ограничения общности можно считать, что $f_1^\ast(x_1) \equiv 0$ и $f_2^\ast(x_2) \equiv 0$, в противном случае мы внесем их внутрь $f_{12}^\ast(x_1, x_2)$.

$\mathbb{E}\ \mathcal{L}(y, f_{12}^\ast(x_1, x_2)) < \mathbb{E}\ \mathcal{L}(y, f_1(x_1) + f_2(x_2) + f_{12}(x_1, x_2)) \tag{4}$

Введем следующее обозначение:

$f_{12}^{\ast\ast}(x_1, x_2) = f_{12}^\ast(x_1, x_2) - f_1(x_1) - f_2(x_2)$

Подставим $f_{12}^{\ast\ast}$ вместо $f_{12}$ в уравнение $(3)$. Поскольку верно $(4)$, то мат. ожидание по первому слагаемому в $(3)$ станет ниже, а мат. ожидание по второму слагаемому не изменится, так как мы не меняли $f_1, f_2$. Это говорит о том, что тройка функций $f_1, f_2, f_{12}$ не являлась оптимальной с точки зрения уравнения $(3)$, поскольку тройка функций $f_1, f_2, f_{12}^{\ast\ast}$ дает меньшее значение мат. ожидания.

Таким образом, предположив, что $(1)$ не выполняется, мы пришли к противоречию с предположением об оптимальности тройки $f_1, f_2, f_{12}$, следовательно, $(1)$ выполняется.

**Теорема**.

Если тройка функций $f_1, f_2, f_{12}$ удовлетворяет уравнению $(3)$, то она удовлетворяет также уравнению $(2)$.

**Доказательство**.

Пойдем аналогичным путем. Пусть это не так, то есть существуют более оптимальные $f_1^\ast, f_2^\ast$ с точки зрения условия в уравнении $(2)$:

$\mathbb{E}\ \mathcal{L}(y, f_1^\ast(x_1) + f_2^\ast(x_2)) < \mathbb{E}\ \mathcal{L}(y, f_1(x_1) + f_2(x_2)) \tag{5}$

Введем следующее обозначение:

$f_{12}^{\ast}(x_1, x_2) = f_{12}(x_1, x_2) - f_1^\ast(x_1) - f_2^\ast(x_2) + f_1(x_1) + f_2(x_2)$

Подставив тройку $f_1^\ast, f_2^\ast, f_{12}^{\ast}$ вместо тройки $f_1, f_2, f_{12}$ в уравнение $(3)$ мы получим, что мат. ожидание по первому слагаемому в уравнении $(3)$ не изменилось, а мат. ожидание по второму слагаемому уменьшилось (исходя из $(5)$). Это говорит о том, что тройка функций $f_1, f_2, f_{12}$ не являлась оптимальной с точки зрения уравнения $(3)$, поскольку тройка функций $f_1^\ast, f_2^\ast, f_{12}^\ast$ дает меньшее значение мат. ожидания.

Таким образом, предположив, что $(2)$ не выполняется, мы пришли к противоречию с предположением об оптимальности тройки $f_1, f_2, f_{12}$, следовательно, $(2)$ выполняется.
